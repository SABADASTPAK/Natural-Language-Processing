```python
import pandas as pd
```

```python
df = pd.read_csv("tmdb_5000_movies.csv")
df.head()
df.columns 
```

```python
df["combined"] = df["original_title"] + " " + df["overview"] + " " + df["genres"] + " " + df["keywords"] + " " + df["spoken_languages"] + " " + df["title"] +" " + df["status"] +" " + df["tagline"]
df["combined"] = df["combined"].fillna("")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
vectorized_movie_data= tfidf.fit_transform(df["combined"])

```

```python
from sklearn.metrics.pairwise import cosine_similarity
if query not in df['title'].values:
    print("movie does not exist in the database")
else:
    movie_index=df.index[df["title"]==query][0]

    movie_vector=vectorized_movie_data[movie_index]

    similarity_scores=cosine_similarity(movie_vector, vectorized_movie_data).flatten()

    sorted=similarity_scores.argsort()[::-1]
    similar_movie_indices = sorted[1:6]
    
    print("\nTop 5 recommended movies:")
    for idx in sorted:
        print(df["title"].iloc[idx])
```

```python



def get_movie_by_title(df, query):
    # find the row where the title matches
    row = df[df["title"] == query]
    if row.empty:
        return None  # not needed but safe
    return row.index[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def recommend(df,vectorized_movie_data,query,top_n=5):
    idx=get_movie_by_title(df, query)
    if idx is None:
        print("movie not found")
        return
    query_vector=vectorized_movie_data[idx]
    similarities= cosine_similarity(query_vector,vectorized_movie_data).flatten()
    # Add to dataframe temporarily
    df["similarity"] = similarities
    
    # Remove the query movie itself
    results = df[df.index != idx]
    
    # Sort and show top results
    return results.sort_values("similarity", ascending=False)[["title","similarity"]].head(top_n)
```
